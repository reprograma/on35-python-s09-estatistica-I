```python

import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('vendas_ficticias.csv')
```

```python
df.head(50)
```

```python
# Calcular estatísticas descritivas para o valor da venda
valor_venda_stats = {
    'Média': df['Valor da venda'].mean(),
    'Mediana': df['Valor da venda'].median(),
    'Mínimo': df['Valor da venda'].min(),
    'Máximo': df['Valor da venda'].max(),
    'Desvio Padrão': df['Valor da venda'].std()
}

# Calcular estatísticas descritivas para a quantidade vendida
quantidade_vendida_stats = {
    'Média': df['Quantidade vendida'].mean(),
    'Mediana': df['Quantidade vendida'].median(),
    'Mínimo': df['Quantidade vendida'].min(),
    'Máximo': df['Quantidade vendida'].max(),
    'Desvio Padrão': df['Quantidade vendida'].std()
}

valor_venda_stats, quantidade_vendida_stats
```

```python
# Identificar o produto mais vendido
produto_mais_vendido = df['Produto'].mode()[0]

produto_mais_vendido
```

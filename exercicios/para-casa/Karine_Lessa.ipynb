```python
import pandas as pd 
```

```python
df = pd.read_csv('vendas_ficticias.csv')
df
```

```python
## Perguntas para responder
#1.	Calcular a média, mediana, mínimo, máximo e desvio padrão dos valores das vendas e quantidade vendida.

#Cálculo média:
media_valor_venda = df['Valor da venda'].mean()
print(f"A média do valor da venda é: {media_valor_venda}")

#Cálculo mediana:

mediana_valor_venda = df['Valor da venda'].median()
print(f"A mediana do valor da venda é: {mediana_valor_venda}")


#Cálculo mínimo 
minimo_vendas = df['Valor da venda'].min()
print (f"O mínimo do valor das vendas é:{minimo_vendas}")

#Cálculo máximo
maximo_vendas = df['Valor da venda'].max()
print(f"O máximo do valor das vendas é: {maximo_vendas}")

#Cálculo desvio padrão

desvio_padrao = df['Valor da venda'].std()
print(f"O desvio padrão do valor das vendas é: {desvio_padrao}")


#média quantidade vendida

quant_media_vend = df['Quantidade vendida'].mean()
print(f"A quantidade média vendida é: {quant_media_vend}")

#mediana quantidade vendida
mediana_quant_vend = df['Quantidade vendida'].median()
print(f"O valor da mediana da quantidade vendida é:{mediana_quant_vend}")

#Mínimo quantidade vendida
minimo_quant_vend = df['Quantidade vendida'].min()
print(f"O mínimo da quantidade vendida é:  {minimo_quant_vend}")

#Máximo quantidade vendida
maximo_quant_vend = df['Quantidade vendida'].max()
print(f"O máximo da quantidade vendida é: {maximo_quant_vend}")

#Desvio padrão quantidade vendida
desvio_padrao_quant_vend = df['Quantidade vendida'].std()
print(f"O desvio padrão da quantidade vendida é: {desvio_padrao_quant_vend}")

# 2. Qual produto que mais vendeu? (Usar a moda para identificar).

mais_vendas = df['Produto'].mode()
print(f"O produto mais vendido foi: {mais_vendas}")
```

### Conjunto de Dados
Os dados estão disponíveis em formato csv (‘vendas_ficticias.csv’) e contêm as seguintes informações para cada venda:

Data da venda / Valor da venda / Vendedor / Quantidade vendida / Produto

### Perguntas para responder
1.	Calcular a média, mediana, mínimo, máximo e desvio padrão dos valores das vendas e quantidade vendida.
2.	Qual produto que mais vendeu? (Usar a moda para identificar).

```python
#import bibiliotecas
import pandas as pd
from tabulate import tabulate
```

```python
df = pd.read_csv('vendas_ficticias.csv') #leitura do arquivo csv
```

```python
print(tabulate(df, headers='keys', tablefmt='fancy_grid')) #print tabela
```

```python
def calculo_estatistico():
    # Calculando valor das vendas
    media_valor = df['Valor da venda'].mean()
    mediana_valor = df['Valor da venda'].median()
    minimo_valor = df['Valor da venda'].min()
    maximo_valor = df['Valor da venda'].max()
    desvio_padrao_valor = df['Valor da venda'].std()

    # Exibir os resultados de Valor de Venda
    print("Valor da Venda:")
    print(f"Media: {media_valor}")
    print(f"Mediana: {mediana_valor}")
    print(f"Minimo: {minimo_valor}")
    print(f"Maximo: {maximo_valor}")
    print(f"Desvio Padrao: {desvio_padrao_valor}\n")

    # Calculando a quantidade vendida
    media_qtde = df['Quantidade vendida'].mean()
    mediana_qtde = df['Quantidade vendida'].median()
    minimo_qtde = df['Quantidade vendida'].min()
    maximo_qtde = df['Quantidade vendida'].max()
    desvio_padrao_qtde = df['Quantidade vendida'].std()

    # Exibir os resultados de Quantidade de Venda
    print("Quantidade Vendida:")
    print(f"Media: {media_qtde}")
    print(f"Mediana: {mediana_qtde}")
    print(f"Minimo: {minimo_qtde}")
    print(f"Maximo: {maximo_qtde}")
    print(f"Desvio Padrao: {desvio_padrao_qtde}\n")

    # Calculando a moda de Produto
    prod_mais_vendido = df['Produto'].mode()[0]

    # Exibir o resultado de Produto mais vendido
    print(f"O produto que mais vendeu foi: {prod_mais_vendido}")
```

```python
calculo_estatistico()
```
